```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ---------- Settings ----------
SAMPLES_PER_DAY = 144
USE_OBSERVED_DAILY_TOTAL = True
FAILURE_WINDOW = ("2021-07-26", "2021-07-30")
PCT_POINTS = np.arange(0, 101, 1)
APPLY_DEBOUNCE_2OF_3 = False
SHOW_P95_REFERENCE = True
SHOW_PICK_LINE = True

# Colors
COLOR_FP       = "tab:blue"
COLOR_STRESSED = "tab:orange"
COLOR_P95      = "crimson"
COLOR_PICK     = "tab:green"

# ---------- Helpers ----------
def daily_from_class_error(df, use_observed=True, samples_per_day=144):
    ce = pd.to_numeric(df["class_error"], errors="coerce").fillna(0).astype(int)
    if APPLY_DEBOUNCE_2OF_3:
        ce = (ce.rolling(3, center=True).sum() >= 2).astype(int)
    daily_cnt = ce.resample("D").sum()
    if use_observed:
        daily_tot = ce.resample("D").count()
        daily_pct = (daily_cnt / daily_tot.where(daily_tot > 0)) * 100.0
    else:
        daily_pct = (daily_cnt / samples_per_day) * 100.0
    return daily_cnt.rename("alerts"), daily_pct.rename("anomaly_pct")

def sweep_anomaly_only(daily_cnt, daily_pct, failure_window, pct_points):
    exclude = pd.date_range(failure_window[0], failure_window[1], freq="D")
    baseline = daily_pct[~daily_pct.index.isin(exclude)].dropna()

    thr_vals, fp_pct, stressed_pct = [], [], []
    den = daily_cnt.sum()

    for p in pct_points:
        thr = float(baseline.quantile(p/100.0)) if len(baseline) else np.nan
        thr_vals.append(thr)
        m = daily_pct >= thr
        stressed_pct.append(float(m.mean() * 100.0))
        fp = (daily_cnt[~m].sum() / den) if den > 0 else np.nan
        fp_pct.append(float(fp * 100.0))

    thr_vals, fp_pct, stressed_pct = map(np.asarray, (thr_vals, fp_pct, stressed_pct))

    # intersection or closest approach
    diff = fp_pct - stressed_pct
    idx = np.where(np.sign(diff[:-1]) * np.sign(diff[1:]) < 0)[0]
    if idx.size:
        i = idx[0]
        x0, x1, y0, y1 = pct_points[i], pct_points[i+1], diff[i], diff[i+1]
        x_star = x0 + (0 - y0) * (x1 - x0) / (y1 - y0)
    else:
        i = int(np.nanargmin(np.abs(diff)))
        x_star = float(pct_points[i])

    ix = int(round(x_star))
    return {
        "x": pct_points,
        "thr_vals": thr_vals,
        "fp_pct": fp_pct,
        "stressed_pct": stressed_pct,
        "x_star": x_star,
        "thr_star": float(thr_vals[ix]),
        "fp_star": float(fp_pct[ix]),
        "cov_star": float(stressed_pct[ix]),
    }


def nice_ceiling(x, base=5):
    return base * int(np.ceil(x / base)) if np.isfinite(x) else base

# ---------- Precompute all results & global right-axis limit ----------
all_res = []
global_max_stressed = 0.0

for df, meta in zip(dataframe_lst, data_path_lst):
    d = df.copy().sort_index()
    if not isinstance(d.index, pd.DatetimeIndex):
        d.index = pd.to_datetime(d.index)
    daily_cnt, daily_pct = daily_from_class_error(
        d, use_observed=USE_OBSERVED_DAILY_TOTAL, samples_per_day=SAMPLES_PER_DAY
    )
    res = sweep_anomaly_only(daily_cnt, daily_pct, FAILURE_WINDOW, PCT_POINTS)
    res["title"] = meta["experiment"]
    all_res.append(res)
    mx = np.nanmax(res["stressed_pct"])
    if np.isfinite(mx):
        global_max_stressed = max(global_max_stressed, mx)

# RIGHT_Y_MAX = nice_ceiling(global_max_stressed, base=5)  # same range for all right y-axes

RIGHT_Y_MAX = 100  # same range for all right y-axes


# ---------- Plot 3×3 grid ----------
rows, cols = 3, 3
FS_TITLE, FS_LAB, FS_TICK, LW = 16, 12, 10, 2.0
plt.rcParams.update({
    "figure.figsize": (18, 12),
    "axes.titlesize": FS_TITLE,
    "axes.labelsize": FS_LAB,
    "xtick.labelsize": FS_TICK,
    "ytick.labelsize": FS_TICK,
    "lines.linewidth": LW,
})

fig, axes = plt.subplots(rows, cols, sharex=True)
axes = np.asarray(axes).reshape(rows, cols)

for k, res in enumerate(all_res):
    r, c = divmod(k, cols)
    ax1 = axes[r, c]

    # Left axis: FP
    ax1.plot(res["x"], res["fp_pct"], color=COLOR_FP, label="Approx. false-positive fraction")
    ax1.set_ylabel("FP (%)", color=COLOR_FP)
    ax1.tick_params(axis='y', colors=COLOR_FP)
    ax1.spines['left'].set_color(COLOR_FP)
    ax1.grid(True, which="both", alpha=0.25)
    ax1.set_ylim(0, 100)            # <<=== unified right y-axis


    # Right axis: % days stressed (with SAME LIMITS everywhere)
    ax2 = ax1.twinx()
    ax2.plot(res["x"], res["stressed_pct"], "--", color=COLOR_STRESSED, label="Days labeled stressed")
    ax2.set_ylabel("Days stressed (%)", color=COLOR_STRESSED)
    ax2.tick_params(axis='y', colors=COLOR_STRESSED)
    ax2.spines['right'].set_color(COLOR_STRESSED)
    ax2.set_ylim(0, RIGHT_Y_MAX)            # <<=== unified right y-axis
    ax2.set_yticks(np.linspace(0, RIGHT_Y_MAX, 4))  # tidy ticks (optional)

    # # P95 reference
    # if SHOW_P95_REFERENCE:
    #     ax1.axvline(95, color=COLOR_P95, linestyle=":", linewidth=1.6, alpha=0.8)

    # # Pick marker
    # if SHOW_PICK_LINE:
    #     ax1.axvline(res["x_star"], color=COLOR_PICK, linestyle="--", alpha=0.8)
    #     ax1.scatter([res["x_star"]], [res["fp_star"]], color=COLOR_PICK, s=18, zorder=5)
    #     ax2.scatter([res["x_star"]], [res["cov_star"]], color=COLOR_PICK, s=18, zorder=5)

    ax1.set_title(res["title"])
    if r == rows - 1:
        ax1.set_xlabel("Baseline percentile of daily anomaly rate (day threshold)")
    if c != 0:
        ax1.set_ylabel("")
    if c != cols - 1:
        ax2.set_ylabel("")

# Global legend
handles = [
    plt.Line2D([0],[0], color=COLOR_FP, lw=LW, label="Approx. false-positive fraction"),
    plt.Line2D([0],[0], color=COLOR_STRESSED, lw=LW, linestyle="--", label="Days labeled stressed"),
]
fig.legend(handles=handles, loc="upper left", ncol=2, framealpha=0.95, borderaxespad=0.6, fontsize=FS_LAB)
fig.suptitle("", y=0.98)

fig.tight_layout(rect=[0, 0, 1, 0.95])
pdf_name = "anomaly_only_fp_vs_percentile_grid_3x3_uniform_right_axis.pdf"
plt.savefig(pdf_name, bbox_inches="tight")
plt.show()
print("Saved:", pdf_name)
```

```python
import numpy as np
import pandas as pd

# ---- knobs you can tune ----
SAMPLES_PER_DAY = 144                 # 10-min cadence
USE_OBSERVED_DAILY_TOTAL = True       # True: handle missing timestamps; False: assume 144
FAILURE_WINDOW = ("2021-07-26", "2021-07-30")  # exclude from baseline when computing P90
APPLY_DEBOUNCE_2OF3 = False           # set True to remove singletons
INCLUDE_EQUAL = False                 # False: stressed if daily% > P90; True: >= P90
PERCENTILE = 0.90                     # ← P90

rows = []
for df, meta in zip(dataframe_lst, data_path_lst):
    d = df.copy().sort_index()
    if not isinstance(d.index, pd.DatetimeIndex):
        d.index = pd.to_datetime(d.index)

    # class_error → daily counts & daily anomaly %
    ce = pd.to_numeric(d["class_error"], errors="coerce").fillna(0).astype(int)
    if APPLY_DEBOUNCE_2OF3:
        ce = (ce.rolling(3, center=True).sum() >= 2).astype(int)

    daily_cnt = ce.resample("D").sum()
    if USE_OBSERVED_DAILY_TOTAL:
        daily_tot = ce.resample("D").count()
        daily_pct = (daily_cnt / daily_tot.where(daily_tot > 0)) * 100.0
    else:
        daily_pct = (daily_cnt / SAMPLES_PER_DAY) * 100.0
    daily_pct = daily_pct.rename("anomaly_pct")

    # baseline (exclude the failure window) and compute P90
    exclude = pd.date_range(FAILURE_WINDOW[0], FAILURE_WINDOW[1], freq="D")
    baseline_pct = daily_pct[~daily_pct.index.isin(exclude)].dropna()
    P90 = float(baseline_pct.quantile(PERCENTILE)) if len(baseline_pct) else np.nan

    # metrics at P90
    if np.isfinite(P90) and daily_cnt.sum() > 0:
        stressed_mask = (daily_pct >= P90) if INCLUDE_EQUAL else (daily_pct > P90)
        approx_fp = (daily_cnt[~stressed_mask].sum() / daily_cnt.sum()) * 100.0
        days_stressed_pct = stressed_mask.mean() * 100.0
    else:
        approx_fp = np.nan
        days_stressed_pct = np.nan

    rows.append({
        "experiment": meta["experiment"],
        "p90_threshold_daily_anomaly_%": P90,
        "approx_fp_at_p90_%": approx_fp,
        "days_stressed_at_p90_%": days_stressed_pct,
        "n_days": int(daily_cnt.notna().sum()),
    })

p90_table = pd.DataFrame(rows)
print(p90_table)

# Save for the manuscript / rebuttal
p90_table.to_csv("p90_thresholds_all_experiments.csv", index=False, float_format="%.4f")
print("Saved: p90_thresholds_all_experiments.csv")
```
