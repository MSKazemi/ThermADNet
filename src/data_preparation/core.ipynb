## Create daily average for core temp

```python
%reset -f
import sys, os
sys.path.append('/home/seyedkazemi/codes/mskhelper/')
import pandas as pd, numpy as np
import dask.dataframe as ds
import random
import logging
import datetime, pytz
import mohsenutils
import imp 
import m100_preprocessing_helper
import data_provider
imp.reload(m100_preprocessing_helper)


def mean_multp_pd(list_pd_df):
    return pd.concat(list_pd_df).groupby(pd.concat(list_pd_df).index).mean()    


def _lst_datafram_same_day(path, pub, metric_lst, first_metric, first_file):
    kwargs = {'date_parser':pd.to_datetime,  'parse_dates':True,  'index_col':'timestamp'}
    dataframes = []
#     dataframes = [pd.read_csv(os.path.join(path, pub, first_metric, first_metric+'_pivot',first_file), **kwargs)]
    
    processed_file_name_path_lst = []
#     processed_file_name_path_lst = [[path, pub, first_metric, first_metric+'_pivot',first_file]]
    
    date = first_file.split(first_metric)[1][1:].split('(')[0].split('.csv')[0]
    print('Date: ',date, ' ', 25*'-+-+')
    
    for second_metric in metric_lst:
        second_file_lst = [f for f in os.listdir(os.path.join(path, pub, second_metric, second_metric+'_pivot')) if date in f]
        print('second_file_lst: ', second_file_lst, '  len(second_file_lst):', len(second_file_lst))
        try:
            for second_file in second_file_lst:
                dataframes.append(pd.read_csv(os.path.join(path, pub, second_metric, second_metric+'_pivot',second_file), **kwargs))
                processed_file_name_path_lst.append([path, pub, second_metric, second_metric+'_pivot',second_file])
                print('second_file ',second_file,' appended')
        except Exception as e:
            print(e)
    
    print('First_metric',first_metric, first_file, date,'len dataframes', len(dataframes))
    print()
    return dataframes, processed_file_name_path_lst


def files_transfer(source_lst):
    
    '''
    source_lst = [['/home/seyedkazemi/dataset/Marconi100_backup/', 'ipmi_pub', 'p1_core1_temp', 'p1_core1_temp_pivot', 'p1_core1_temp_2021-04-08_00-00-00_to_2021-04-09_00-00-00.csv.gz'], 
                  ['/home/seyedkazemi/dataset/Marconi100_backup/', 'ipmi_pub', 'p1_core3_temp', 'p1_core3_temp_pivot', 'p1_core3_temp_2021-04-08_00-00-00_to_2021-04-09_00-00-00.csv.gz'], 
                  ['/home/seyedkazemi/dataset/Marconi100_backup/', 'ipmi_pub', 'p1_core2_temp', 'p1_core2_temp_pivot', 'p1_core2_temp_2021-04-08_00-00-00_to_2021-04-09_00-00-00.csv.gz']]
    '''
    
    for path in source_lst:
        
        dst_path = os.path.join(*path[:-2],path[-2]+'_done','')
        
        if not os.path.isdir(dst_path):
            os.makedirs(     dst_path)
            print(           dst_path, 'created !!!')
        os.replace(os.path.join(*path), os.path.join(dst_path, path[-1]))
        
def create_av_new_metric_all_days(path, pub, metric_lst, new_mtric_name):
    random.shuffle(metric_lst)
    first_metric = metric_lst[0]
#     metric_lst.remove(first_metric)
    files_first_metric = [f for f in os.listdir('/home/seyedkazemi/dataset/Marconi100_backup/ipmi_pub/' + first_metric + '/' + first_metric + '_pivot/') if '.csv.gz' in f]
    if len(files_first_metric)>0:
        
        path_new_mtric_name_pivot = os.path.join(path, pub, new_mtric_name, new_mtric_name+'_pivot','')
        
        if not os.path.isdir(path_new_mtric_name_pivot):
            os.mkdir(        path_new_mtric_name_pivot)

        for first_file in files_first_metric:
            dataframes, processed_file_name_path_lst = _lst_datafram_same_day(path, pub, metric_lst, first_metric, first_file)
            mean_data_one_day = mean_multp_pd(dataframes)

            date = first_file.split(first_metric)[1][1:].split('(')[0].split('.csv')[0]

            mean_data_one_day.to_csv(path_new_mtric_name_pivot + new_mtric_name + '_' + date + '.csv.gz')
            print('len processed_file_name_path_lst', len(processed_file_name_path_lst), processed_file_name_path_lst)
            files_transfer(processed_file_name_path_lst)

        

        

        


path = '/home/seyedkazemi/dataset/Marconi100_backup/'
pub = 'ipmi_pub'
metric_lst_core0 = ['p0_core0_temp','p0_core10_temp','p0_core11_temp','p0_core12_temp','p0_core13_temp','p0_core14_temp','p0_core15_temp','p0_core16_temp',
                   'p0_core17_temp','p0_core18_temp','p0_core19_temp','p0_core1_temp','p0_core20_temp','p0_core21_temp','p0_core22_temp','p0_core23_temp',
                   'p0_core2_temp','p0_core3_temp','p0_core4_temp','p0_core5_temp','p0_core6_temp','p0_core7_temp','p0_core8_temp','p0_core9_temp']

metric_lst_core1 = ['p1_core0_temp','p1_core10_temp','p1_core11_temp','p1_core12_temp','p1_core13_temp','p1_core14_temp','p1_core15_temp','p1_core16_temp',
                   'p1_core17_temp','p1_core18_temp','p1_core19_temp','p1_core1_temp','p1_core20_temp','p1_core21_temp','p1_core22_temp','p1_core23_temp',
                   'p1_core2_temp','p1_core3_temp','p1_core4_temp','p1_core5_temp','p1_core6_temp','p1_core7_temp','p1_core8_temp','p1_core9_temp']


metric_lst_gpu =     ['gpu0_core_temp','gpu1_core_temp','gpu3_core_temp','gpu4_core_temp']
metric_lst_gpu_mem = ['gpu0_mem_temp','gpu1_mem_temp','gpu3_mem_temp','gpu4_mem_temp']
 
create_av_new_metric_all_days(path=path, pub=pub, metric_lst=metric_lst_core0, new_mtric_name= 'cores_temp0')
create_av_new_metric_all_days(path=path, pub=pub, metric_lst=metric_lst_core1, new_mtric_name= 'cores_temp1')
create_av_new_metric_all_days(path=path, pub=pub, metric_lst=metric_lst_gpu, new_mtric_name= 'gpu_temp')
create_av_new_metric_all_days(path=path, pub=pub, metric_lst=metric_lst_gpu_mem, new_mtric_name= 'gpu_mem_temp')
```
